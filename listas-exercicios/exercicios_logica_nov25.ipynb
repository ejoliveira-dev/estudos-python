```python
# --- Exercícios 1 ---
# -> Variáveis.

# Exercício 1.1: Calculadora de Consumo de Energia.
# Crie um programa que peça ao usuário a potência de um equipamento (em Watts), o número de horas que ele fica ligado por dia e o valor do kWh (em reais). O programa deve calcular e exibir o custo mensal (considerando 30 dias) desse equipamento.

potencia = float(input("Escreva a potência do seu equipamento (em Watts): "))
horas_ligado_dia = float(input("Escreva o tempo (em horas por dia) em que ele fica ligado: "))
valor_kwh = float(input("Escreva o valor do quilowatt-hora (em reais): "))

consumo_mensal = (potencia / 1000) * horas_ligado_dia * 30 # primeiro transformamos watts em kwh, depois pelas horas e pela quantidade de dias em um mês.
custo_mensal = consumo_mensal * valor_kwh

print(f"\nO custo mensal desse equipamento é de: R${custo_mensal}.")
```

```python
# Exercício 1.2: Conversor de Temperatura.
# Crie um script que solicite uma temperatura em graus Celsius e a converta para Fahrenheit. Exiba o resultado formatado.

temp_celsius = float(input("Adicione a temperatura em graus Celsius: "))

fahrenheit = (temp_celsius * 1.8) + 32

print(f"\n{temp_celsius}ºC é igual a {fahrenheit}ºF em Fahrenheit!")
```

```python
# --- Exercícios 2 ---
# -> Estruturas Condicionais (if, elif, else).

# Exercício 2.1: Verificação de Aprovação.
# Peça duas notas de um aluno (N1 e N2). Calcule a média aritmética.

n1 = float(input())
n2 = float(input())

media = (n1 + n2)/2

if media >= 7.0:
  print("\nAprovado!")
elif media >= 4.0 and media < 7.0:
  print("\nRecuperação.")
else:
  print("\nReprovado.")
```

```python
# Exercício 2.2: Classificação de Tensão (Contexto Engenharia).
# Solicite ao usuário o valor de uma tensão elétrica (Volts).

tensao_elet = int(input("Digite a tensão elétrica (V): "))

if tensao_elet <= 1000: # uma tensão igual ou menor que 1000 é baixa.
  print(f"\n{tensao_elet}V é baixa tensão.")
elif tensao_elet > 1000 and tensao_elet <= 36000: # uma maior que 1000 e menor ou igual a 36000 é média.
  print(f"\n{tensao_elet}V é média tensão.")
elif tensao_elet > 36000: # e a maior que 36000 é alta.
  print(f"\n{tensao_elet}V é alta tensão.")
```

```python
# --- Exercícios 3 ---
# -> Estruturas de Repetição (for, while).

# Exercício 3.1: Tabuada Dinâmica.
# Peça um número inteiro ao usuário e utilize um loop for para imprimir a tabuada desse número de 1 a 10.

numero = int(input("Qual a tabuada desejada?: "))

for i in range(1, 10): # para percorrer de 1 a 10.
  resultado = numero * (i + 1) # resultado é igual o numero vezes o contador + 1.
  print(f"\n{numero} x {i + 1} = {resultado}.") # formatação da tabuada.
```

```python
# Exercício 3.2: Somatório de Pares.
# Utilize um loop while ou for para somar todos os números pares entre 1 e 100. Exiba apenas o resultado final da soma.

i = 1 # contador, vai ajudar a "percorrer" os números/contar.
soma = 0 # acumulador começa em zero para poder ser incrementado a cada loop.

while i <= 100: # enquanto o contador for menor ou igual a 100, o loop roda.
  if i % 2 == 0: # se for um número par, o contador é adicionado ao acumulador.
     soma = soma + i
  i = i + 1 # aqui é pro contador continuar aumentando.

print(f"A soma de todos os pares é {soma}!")

```

```python
# --- Exercícios 4 ---
# -> Listas e Manipulação Básica de Dados.

# Exercício 4.1: Lista de Equipamentos.
# Crie uma lista vazia chamada equipamentos.
# Use um loop para pedir ao usuário o nome de 3 equipamentos elétricos e adicione-os à lista (use append).

lista_equip = [] # lista vazia

while len(lista_equip) < 3: # enquanto a quantidade de elementos for menor que 3, continue a perguntar
  equipamento = str(input("Adicione um equipamento: "))
  lista_equip.append(equipamento) # adiciona a lista

print(f"\n{lista_equip}\n") # imprime
lista_equip[1] # imprime a segunda posição, no caso o índice 1.
```

```python
# Exercício 4.2: Análise de Temperaturas.
# Dada a lista de temperaturas medidas em um dia: temps = [22.5, 24.0, 26.5, 29.0, 23.5, 21.0].
# Escreva um código que calcule e exiba:
# A temperatura máxima, a temperatura mínima, a temperatura média do dia.

temperaturas = [22.5, 24.0, 26.5, 29.0, 23.5, 21.0]

temperaturas.sort() # função sort pra deixar a lista em ordem permanentemente.

print(f"A temperatura máxima do dia foi de {max(temperaturas)}.") # funções max e min pra pode declarar o maior e o menor elemento.
print(f"A temperatura mínima do dia foi de {min(temperaturas)}.")

soma = 0
for temperatura in temperaturas: # laço para percorrer e somar as temperaturas.
  soma = soma + temperatura

media = soma // len(temperaturas) # dividindo a soma pela quantidade de elementos.

print(f"A temperatura média do dia foi {media}.")
```
